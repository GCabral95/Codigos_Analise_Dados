```python
import pandas as pd
import geopy.distance
```

```python
df = pd.read_excel('geo.xlsx')
```

```python
df.head()
```

```python
dist = pd.DataFrame(columns=['Ponto1','Ponto2','distancia'])
for x,y,z in zip(df['LAT'],df['LONG'],df['Ponto']):
    for a,b,c in zip(df['LAT'],df['LONG'],df['Ponto']):
        coord1 = (x,y)
        coord2 = (a,b)
        D = geopy.distance.geodesic(coord1, coord2).km
        dist = dist.append({'Ponto1' : z, 'Ponto2' : c,'distancia': D}, ignore_index=True)
```
