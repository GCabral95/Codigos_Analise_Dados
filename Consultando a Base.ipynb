```python
#Agregar dados por varias funcoes
df.groupby('type1').sp_attack.agg(['mean','max']) #maior e media de especial ataque por tipo
```

```python
#Agregar dados por varias colunas e funcoes com uma variavel numerica
df.groupby(['type1','type2']).sp_attack.agg(['mean','max']) #maior e media de especial ataque por tipo
```

```python
#Agregar os dados por multiplas colunas numericas
df.groupby('type1').agg({'speed':['mean','max'], #mÃ©dia e maximo velocidade
                         'sp_defense': ['mean','min']}) #media e minimo defesa especial
```

```python
#Agrupando a partir de associaÃ§ao de colunas
habilidades_tot = {'sp_attack': 'ataque', 'attack': 'ataque', 'sp_defense': 'defesa', 'defense': 'defesa'}
```

```python
df2 = df[['name','sp_attack','attack','sp_defense','defense']]
df2.set_index('name', inplace=True)
```

```python
df2.head()
```

```python
(df2.groupby(habilidades_tot, axis=1).sum()).head()
```

```python
#Agrupando por uma funÃ§Ã£o criada
def dif_quantil(x):
    return (x.quantile(0.9) - x.quantile(0.1))
```

```python
df.groupby('type1')[['attack','defense']].agg(dif_quantil)
```

```python
#Agrupando por fraÃ§Ãµes da base
faixas_peso = [0,250,500,750,1000]
grupos = pd.cut(df['weight_kg'],faixas_peso)
```

```python
df.groupby(grupos)[['attack','sp_attack']].mean()
```

```python
#Agrupando por quantil
grupos = pd.qcut(df['weight_kg'],[0, 0.2,0.5,0.8,1])
```

```python
df.groupby(grupos)[['attack','sp_attack']].mean()
```
