```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_excel('base_2.xlsx')
```

```python
df.head()
```

# AB 1

```python
df_1 = df[df['Ab'] == 1]
```

```python
df_1_a = df_1[df_1['P'].isin([1,4])]
df_1_b = df_1[~df_1['P'].isin([1,4])]
df_1_a = df_1_a[['X','N','T']]
df_1_b = df_1_b[['X','N','T']]
df_ufs = df_1[['X','Y']]
df_ufs = df_ufs.drop_duplicates()
```

# AB1_ag

```python
df_1_a.head()
```

```python
df_1_a_count = df_1_a.pivot_table(index = ['X'], columns = ['T'], aggfunc = 'count').reset_index()
```

```python
df_1_a_count.columns = list(map("_".join, df_1_a_count.columns))
df_1_a_count = df_1_a_count.fillna(0)
df_1_a_count= df_1_a_count.rename(columns = {'X_':'X'})
df_1_a_count
```

```python
df_1_a_count['total'] = df_1_a_count['N_C'] + df_1_a_count['N_F'] + df_1_a_count['N_R']
```

```python
df_1_a_count = pd.merge(df_1_a_count[['X','total']],df_ufs[['X','Y']], on = 'X', how = 'left')
```

```python
df_1_a_count
```

```python
def reg(x):
    if x=='A' or x=='B':
        return 'G1'
    else:
        return 'G2'
```

```python
df_1_a_count['reg'] = df_1_a_count.apply(lambda x: reg(x.Y), axis=1)
```

```python
df_1_a_count
```

```python
def cl(x):
    if x<3:
        return 'ate 2'
    elif x<6:
        return 'ate 5'
    else:
        return 'maior que 5'
```

```python
df_1_a_count['cl'] = df_1_a_count.apply(lambda x: cl(x.total), axis=1)
```

```python
df_1_a_count
```

```python
def k_1(x):
    if x<3:
        return math.ceil((x*4*1))
    elif x<6:
        return math.ceil((x*4*0.8))
    else:
        return math.ceil((x*4*0.7))
```

```python
df_1_a_count['k_1'] = df_1_a_count.apply(lambda x: k_1(x.total), axis=1)
df_1_a_count['k_2'] = df_1_a_count['total']
df_1_a_count['k_3'] = 4
df_1_a_count['k_4'] = 1
```

```python
df_1_a_count
```

```python
def k_5(x,y):
    if x == 'G1':
        if y == 'ate 2':
            return 1
        elif y == 'ate 5':
            return 2
        else:
            return 3
    else:
        if y == 'ate 2':
            return 2
        elif y == 'ate 5':
            return 3
        else:
            return 4
```

```python
df_1_a_count['k_5'] = df_1_a_count.apply(lambda x: k_5(x.reg,x.cl), axis=1)
```

```python
df_1_a_count
```

```python
def k_6(x,y):
    if x == 'G1':
        if y == 'ate 2':
            return 1
        elif y == 'ate 5':
            return 1
        else:
            return 2
    else:
        if y == 'ate 2':
            return 2
        elif y == 'ate 5':
            return 2
        else:
            return 3
```

```python
df_1_a_count['k_6'] = df_1_a_count.apply(lambda x: k_6(x.reg,x.cl), axis=1)
```
